```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs

  

  
# find out covariance with respect  rows 
```

```
url="https://raw.githubusercontent.com/Anasuya-Sahoo/DMDW-Lab/main/student-mat.csv"
df=pd.read_csv(url)
df.head()
```

```
numerical=df[['age','Medu','Fedu','studytime','traveltime','failures']]
numerical.head()
```

```
num2=np.array(numerical[['traveltime','studytime']])
num2=num2.reshape(-1,2)
num2.shape
```

```
covMatrix = np.cov(num2,bias=False)

  
print("shape of matrix traveltime and studytime:", np.shape(covMatrix )) 
  
print("shape of covariance matrix:", np.shape(np.cov(covMatrix ))) 
  
print(np.cov(covMatrix )) 
```
